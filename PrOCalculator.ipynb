```python
import numpy as np
import pandas as pd
import sys

#PERFORM CALCULATIONS AND CREATE THE DATAFRAME

#Create the Dataframe
filename = 'PrOData.csv'
df = pd.read_csv(filename)

#Calculate Pro%
df['PrO%'] = df['PrO']/df['PrOO']

#Calculate the Average PrO%
tot_prop = df['PrO%'].sum()
tot_rows = len(df)
avg_prop = round(tot_prop/tot_rows, 3)

#Calculate the % of PrO that resulted in RBIs
df['RBI%'] = df['RBI']/df['PrO']

#Calculate the total strike out percentages
df['SO%'] = df['SO']/df['TOT_AB']

#Calculate the strike out percentage in PrO Opportunities
df['PrOSO%'] = df['PrOK']/df['PrOO']
    
def player_lookup(first, last):
    lookup_row = df.loc[(df['LAST_NAME_TX'] == last) & (df['FIRST_NAME_TX'] == first)]
    return lookup_row

df.head()
```

```python
print('######## WELCOME TO THE 2010-2016 Pro% Calculator ########')

def get_input():

    print('Enter the first name of the player: ')
    first = input()
    print('Enter the last name of the player: ')
    last = input()
    
    if((first in df.values) and (last in df.values)):
        row = player_lookup(first, last)
    else:
        print('Player not contained in database, please enter another name')
        get_input()
        
    row_prop = row['PrO%'].to_string(index = False)
    prop_diff = round(abs(float(row_prop) - float(avg_prop)), 3)
    

    print(' ')
    print('Productive Out Statistics for {0} {1}: '.format(first, last))
    print(' ')

    print('PrOO: {0}'.format(round(row['PrOO'], 3).to_string(index = False)))
    print('PrO: {0}'.format(round(row['PrO'], 3).to_string(index = False)))
    print('PrO%: {0}'.format(round(row['PrO%'], 3).to_string(index = False)))
    print('PrO_RBI: {0}'.format(round(row['RBI'], 3).to_string(index = False)))
    print('PrO_RBI %: {0}'.format(round(row['RBI%'], 3).to_string(index = False)))
    print('Total SO %: {0}'.format(round(row['SO%'], 3).to_string(index = False)))
    print('PrO SO%: {0}'.format(round(row['PrOSO%'], 3).to_string(index = False)))

    print(' ')
    if(float(row_prop) < float(avg_prop)):
        print('{0} {1} is {2} below the league average for PrO%'.format(first, last, prop_diff))
    else:
        print('{0} {1} is {2} above the league average for PrO%'.format(first, last, prop_diff))
    print('The league average for PrO% is {0}'.format(avg_prop))
    
    print(' ')
    print('Would you like to look up another player\'s productive out stats? [y]/[n]')
    choice = input()
    if (choice == 'y' or choice == 'yes'):
        print (' ')
        get_input()
    else:
        print(' ')
        print('Thank you!')
        quit()
    

get_input()

#print('PrOO: {0},'.format(row['PrOO'])
```
